```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

```python
data = pd.read_csv("heart_disease_data.csv")
```

```python
data.head()
```

```python
# checking for null values any
data.isnull().sum()
```

```python
# no null values
```

```python
# checking of number of responses of "Yes" and "No" . Higher difference in yes and no means lesser accuracy of data
```

```python
data["target"].value_counts()
```

```python
# no big difference between "Yes" and "No"
```

```python
# separating data into values and targets
```

```python
X = data.drop(columns="target",axis=1);
Y = data["target"]
```

```python
print(X)
```

```python
print(Y)
```

```python
scaler = StandardScaler()
```

```python
# all the column values should be in the same range so for achieving that we are doing preprocessing of data
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)

# 20% of data has been used for testing purpose
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
# training model on Neural Network
```

```python
model = Sequential([
    Dense(32, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(16, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
# Create an instance of the Adam optimizer with a learning rate of 0.001
adam_optimizer = Adam(learning_rate=0.001)
```

```python
# Compile the model
model.compile(optimizer=adam_optimizer,
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
# Train the model
history = model.fit(X_train, Y_train, epochs=200, validation_split=0.2, batch_size=32, verbose=1)
```

```python
# Evaluating the performance of Support Vector Machine Model
```

```python
# training data_prediction
```

```python
test_loss, test_accuracy = model.evaluate(X_test, Y_test)
print(f"Deep Learning Test Accuracy: {test_accuracy}")
```
